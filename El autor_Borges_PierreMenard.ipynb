```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.probability import FreqDist

# Cargar el texto
with open("PierreMenard.txt", "r", encoding="utf-8") as f:
    texto = f.read()

# Tokenizar el texto
tokens = word_tokenize(texto)

# Calcular la frecuencia de las palabras
freq_dist = FreqDist(tokens)

# Obtener las primeras 100 palabras más frecuentes
palabras_mas_frecuentes = freq_dist.most_common(100)

# Imprimir las palabras y sus frecuencias
print("Palabras más frecuentes y sus frecuencias:")
for palabra, frecuencia in palabras_mas_frecuentes:
    print(f"{palabra}: {frecuencia}")

# Calcular las probabilidades de las primeras 100 palabras más frecuentes
total_palabras = sum(freq_dist.values())
probabilidades = {palabra: frecuencia / total_palabras for palabra, frecuencia in palabras_mas_frecuentes}

# Imprimir las probabilidades de las palabras
print("\nProbabilidades de las primeras 100 palabras más frecuentes:")
for palabra, probabilidad in probabilidades.items():
    print(f"{palabra}: {probabilidad:.6f}")
```

```python
import string

# Contar la frecuencia de los signos de puntuación utilizados en el texto
frecuencia_puntuacion = {}
for char in texto:
    if char in string.punctuation:
        if char not in frecuencia_puntuacion:
            frecuencia_puntuacion[char] = 1
        else:
            frecuencia_puntuacion[char] += 1

# Mostrar la frecuencia de los signos de puntuación
print("Frecuencia de los signos de puntuación:")
for char, freq in frecuencia_puntuacion.items():
    print(f"{char}: {freq}")
```

```python
import string
import matplotlib.pyplot as plt

# Contar la frecuencia de los signos de puntuación utilizados en el texto
frecuencia_puntuacion = {}
for char in texto:
    if char in string.punctuation:
        if char not in frecuencia_puntuacion:
            frecuencia_puntuacion[char] = 1
        else:
            frecuencia_puntuacion[char] += 1

# Ordenar la frecuencia de los signos de puntuación por valor descendente
frecuencia_puntuacion = dict(sorted(frecuencia_puntuacion.items(), key=lambda item: item[1], reverse=True))

# Mostrar la frecuencia de los signos de puntuación
print("Frecuencia de los signos de puntuación:")
for char, freq in frecuencia_puntuacion.items():
    print(f"{char}: {freq}")

# Visualizar la frecuencia de los signos de puntuación en un gráfico de barras
plt.figure(figsize=(10, 6))
plt.bar(frecuencia_puntuacion.keys(), frecuencia_puntuacion.values(), color='skyblue')
plt.xlabel('Signos de Puntuación')
plt.ylabel('Frecuencia')
plt.title('Frecuencia de los Signos de Puntuación en el Texto')
plt.xticks(rotation=45)
plt.show()
```
